```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import splitfolders
```

```python
dataGen=ImageDataGenerator(rescale=1)
```

## Deviding Train and Test Data Set

```python
train_dir='D:\\DATA Science\\DEEP LEARNING\\CNN\\CrackDetection\\DataSet\\Crack_by_Split_folder\\train'
test_dir='D:\\DATA Science\\DEEP LEARNING\\CNN\\CrackDetection\\DataSet\\Crack_by_Split_folder\\test'
val_dir='D:\\DATA Science\\DEEP LEARNING\\CNN\\CrackDetection\\DataSet\\Crack_by_Split_folder\\val'
```

```python
train_itt=dataGen.flow_from_directory(train_dir, target_size=(256, 256),batch_size=32)
test_itt=dataGen.flow_from_directory(test_dir, target_size=(256, 256),batch_size=32)
val_itt=dataGen.flow_from_directory(val_dir, target_size=(256, 256),batch_size=32)
```

## Model CNN

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Flatten,Dense
```

```python
model= Sequential()
```

```python
model.add(Conv2D(32,(3,3),activation='relu',input_shape=(256,256,3)))
model.add(MaxPooling2D((2,2)))

model.add(Conv2D(64,(3,3),activation='relu'))
model.add(MaxPooling2D((2,2)))

model.add(Conv2D(128,(3,3),activation='relu'))
model.add(MaxPooling2D((2,2)))

model.add(Flatten())

model.add(Dense(512,activation='relu'))

model.add(Dense(2,activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(train_itt,validation_data=val_itt,epochs=10)
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
plt.plot(losses['loss'])
plt.plot(losses['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('no of epochs')
plt.legend(['train','test'],loc='upper left')
plt.show()
```

```python
plt.plot(losses['accuracy'])
plt.plot(losses['val_accuracy'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('no of epochs')
plt.legend(['train','test'],loc='upper left')
plt.show()
```

```python
model.save('Crack_detection_model.h5')
```
